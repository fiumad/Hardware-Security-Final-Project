### **Task 3: Trojan Analysis & Comparison**
Analyze and compare generated Trojans across different vulnerability types and designs.

```python
#@title Task 3: Trojan Analysis & Comparison
import glob
import re
from collections import defaultdict

def analyze_trojan_characteristics(output_dir="./trojan_outputs"):
    """Comprehensive analysis of generated Trojans"""

    print("🦠 TASK 3: Trojan Analysis & Comparison")
    print("=" * 50)

    if not os.path.exists(output_dir):
        print(f"❌ Output directory {output_dir} not found")
        print("💡 Run Task 2 first to generate Trojaned designs")
        return

    # Find all generated files
    verilog_files = glob.glob(f"{output_dir}/**/*.v", recursive=True)
    taxonomy_files = glob.glob(f"{output_dir}/**/*_taxonomy.txt", recursive=True)

    print(f"📊 Found {len(verilog_files)} Trojaned designs")
    print(f"📋 Found {len(taxonomy_files)} taxonomy files")

    if not verilog_files:
        print("❌ No Trojaned designs found")
        return

    # Analyze Trojan characteristics
    trojan_stats = defaultdict(list)
    vulnerability_analysis = defaultdict(lambda: defaultdict(list))

    for vfile in verilog_files:
        # Extract metadata from filename
        basename = os.path.basename(vfile)
        parts = basename.replace('.v', '').split('_')

        if len(parts) >= 3:
            design_name = parts[0]
            vuln_type = parts[1] if parts[1].startswith('H') else 'Unknown'
            model_name = parts[2] if len(parts) > 2 else 'Unknown'

            # Read Verilog content
            with open(vfile, 'r') as f:
                content = f.read()

            # Extract basic metrics
            lines = len(content.splitlines())
            chars = len(content)

            # Count Verilog constructs
            always_blocks = len(re.findall(r'always\s*@', content))
            if_statements = len(re.findall(r'\bif\s*\(', content))
            registers = len(re.findall(r'\breg\s+', content))

            # Look for suspicious patterns
            hardcoded_vals = len(re.findall(r"\d+'[bh][0-9a-fA-F_xXzZ]+", content))
            counters = len(re.findall(r'<=\s*\w+\s*\+\s*1', content))
            comparisons = len(re.findall(r'==\s*\d+', content))

            stats = {
                'design': design_name,
                'vulnerability': vuln_type,
                'model': model_name,
                'lines': lines,
                'chars': chars,
                'always_blocks': always_blocks,
                'if_statements': if_statements,
                'registers': registers,
                'hardcoded_values': hardcoded_vals,
                'counters': counters,
                'comparisons': comparisons
            }

            trojan_stats[vuln_type].append(stats)
            vulnerability_analysis[vuln_type][design_name].append(stats)

    # Display analysis results
    print("\n" + "=" * 60)
    print("📊 TROJAN CHARACTERISTICS ANALYSIS")
    print("=" * 60)

    for vuln_type, stats_list in trojan_stats.items():
        if not stats_list:
            continue

        print(f"\n🦠 {vuln_type} Trojans ({len(stats_list)} generated):")
        print("-" * 40)

        # Calculate averages
        avg_lines = sum(s['lines'] for s in stats_list) / len(stats_list)
        avg_ifs = sum(s['if_statements'] for s in stats_list) / len(stats_list)
        avg_regs = sum(s['registers'] for s in stats_list) / len(stats_list)
        avg_counters = sum(s['counters'] for s in stats_list) / len(stats_list)

        print(f"   📏 Average lines: {avg_lines:.1f}")
        print(f"   🔀 Average if statements: {avg_ifs:.1f}")
        print(f"   💾 Average registers: {avg_regs:.1f}")
        print(f"   🔢 Average counters: {avg_counters:.1f}")

        # Show sample triggers/payloads from taxonomy files
        sample_taxonomy = None
        for tfile in taxonomy_files:
            if vuln_type.lower() in tfile.lower():
                with open(tfile, 'r') as f:
                    sample_taxonomy = f.read()
                break

        if sample_taxonomy:
            # Extract trigger and payload sections
            trigger_match = re.search(r'Trigger:\s*(.+?)(?:\n\n|Payload:)', sample_taxonomy, re.DOTALL)
            payload_match = re.search(r'Payload:\s*(.+?)(?:\n\n|Taxonomy:)', sample_taxonomy, re.DOTALL)

            if trigger_match:
                trigger = trigger_match.group(1).strip()
                print(f"   🎯 Sample trigger: {trigger[:100]}{'...' if len(trigger) > 100 else ''}")

            if payload_match:
                payload = payload_match.group(1).strip()
                print(f"   💥 Sample payload: {payload[:100]}{'...' if len(payload) > 100 else ''}")

    # Cross-design comparison
    print("\n" + "=" * 60)
    print("🔍 CROSS-DESIGN COMPARISON")
    print("=" * 60)

    design_complexity = defaultdict(list)
    for vuln_type, design_dict in vulnerability_analysis.items():
        for design_name, stats_list in design_dict.items():
            for stats in stats_list:
                complexity = stats['lines'] + stats['if_statements'] * 2 + stats['registers']
                design_complexity[design_name].append(complexity)

    print("\n📈 Design Complexity Rankings:")
    for design, complexities in design_complexity.items():
        avg_complexity = sum(complexities) / len(complexities)
        print(f"   {design}: {avg_complexity:.1f} (avg across {len(complexities)} Trojans)")

    print("\n" + "=" * 60)
    print("💡 ANALYSIS INSIGHTS")
    print("=" * 60)
    print("• Compare Trojan complexity across vulnerability types")
    print("• Identify which designs are most susceptible to Trojans")
    print("• Analyze trigger/payload creativity and sophistication")
    print("• Evaluate stealth characteristics and detectability")
    print("• Consider effectiveness for different attack scenarios")

# Run the analysis
analyze_trojan_characteristics()
```

Validate that generated Trojans are functionally correct and test their trigger conditions.

```python
#@title Task 4: Trojan Validation & Testing
import subprocess
import tempfile
import glob
import os
import re

def validate_trojaned_designs(output_dir="./trojan_outputs"):
    """Validate syntax and basic functionality of generated Trojaned designs"""

    print("🦠 TASK 4: Trojan Validation & Testing")
    print("=" * 50)

    if not os.path.exists(output_dir):
        print(f"❌ Output directory {output_dir} not found")
        print("💡 Run Task 2 first to generate Trojaned designs")
        return

    verilog_files = glob.glob(f"{output_dir}/**/*.v", recursive=True)

    if not verilog_files:
        print("❌ No Trojaned designs found for validation")
        return

    print(f"🔍 Validating {len(verilog_files)} Trojaned designs...")

    validation_results = []

    for vfile in verilog_files[:5]:  # Test first 5 files
        basename = os.path.basename(vfile)
        print(f"\n📋 Validating: {basename}")

        # Read the Trojaned design
        with open(vfile, 'r') as f:
            content = f.read()

        result = {
            'file': basename,
            'syntax_valid': False,
            'has_module': False,
            'has_trojans': False,
            'line_count': len(content.splitlines()),
            'issues': []
        }

        # Basic syntax validation
        try:
            # Check for basic Verilog structure
            if 'module' in content and 'endmodule' in content:
                result['has_module'] = True
                print("   ✅ Module structure found")
            else:
                result['issues'].append("Missing module/endmodule")
                print("   ❌ Missing module structure")

            # Check for Trojan markers
            if 'trojan' in content.lower() or 'TROJANED DESIGN' in content:
                result['has_trojans'] = True
                print("   🦠 Trojan markers detected")

            # Look for suspicious patterns that indicate Trojans
            suspicious_patterns = [
                r"==\s*\d+'[bh][0-9a-fA-F]+",  # Specific value comparisons
                r'<=\s*\w+\s*\+\s*1',         # Counter increments
                r'if\s*\([^)]*==\s*\d+',      # Conditional triggers
            ]

            pattern_count = 0
            for pattern in suspicious_patterns:
                matches = re.findall(pattern, content)
                pattern_count += len(matches)

            if pattern_count > 0:
                print(f"   🔍 Found {pattern_count} suspicious patterns")
                result['has_trojans'] = True

            # Try basic Verilog syntax validation (if iverilog is available)
            try:
                with tempfile.NamedTemporaryFile(mode='w', suffix='.v', delete=False) as tmp:
                    tmp.write(content)
                    tmp.flush()

                    # Try to compile with iverilog if available
                    result_code = subprocess.call(['iverilog', '-t', 'null', tmp.name],
                                                stdout=subprocess.DEVNULL,
                                                stderr=subprocess.DEVNULL)

                    if result_code == 0:
                        result['syntax_valid'] = True
                        print("   ✅ Syntax validation passed")
                    else:
                        result['issues'].append("Syntax errors detected")
                        print("   ⚠️  Syntax validation failed")

                os.unlink(tmp.name)

            except (FileNotFoundError, subprocess.CalledProcessError):
                print("   ℹ️  Syntax validation skipped (iverilog not available)")
                result['syntax_valid'] = None

        except Exception as e:
            result['issues'].append(str(e))
            print(f"   ❌ Validation error: {e}")

        validation_results.append(result)

    # Summary report
    print("\n" + "=" * 60)
    print("📊 VALIDATION SUMMARY")
    print("=" * 60)

    total_files = len(validation_results)
    valid_modules = sum(1 for r in validation_results if r['has_module'])
    valid_syntax = sum(1 for r in validation_results if r['syntax_valid'] is True)
    with_trojans = sum(1 for r in validation_results if r['has_trojans'])

    print(f"📁 Files validated: {total_files}")
    print(f"✅ Valid module structure: {valid_modules}/{total_files}")
    print(f"✅ Syntax validation passed: {valid_syntax}/{sum(1 for r in validation_results if r['syntax_valid'] is not None)}")
    print(f"🦠 Trojan indicators found: {with_trojans}/{total_files}")

    # Show issues if any
    all_issues = []
    for result in validation_results:
        all_issues.extend(result['issues'])

    if all_issues:
        print("\n⚠️  Issues found:")
        for issue in set(all_issues):
            count = all_issues.count(issue)
            print(f"   • {issue} ({count} files)")

    print("\n" + "=" * 60)
    print("🔬 TESTING RECOMMENDATIONS")
    print("=" * 60)
    print("1. 📐 Synthesis Testing:")
    print("   • Use Yosys or Vivado to test synthesis")
    print("   • Check for timing violations or resource usage")
    print("\n2. 🎯 Trigger Testing:")
    print("   • Create testbenches to verify trigger conditions")
    print("   • Test both normal and Trojan-activated behavior")
    print("\n3. 🔍 Detection Testing:")
    print("   • Run static analysis tools on Trojaned designs")
    print("   • Compare with clean versions for differences")
    print("\n4. ⚡ Functional Testing:")
    print("   • Verify original functionality still works")
    print("   • Test edge cases and boundary conditions")

# Run validation
validate_trojaned_designs()
```
